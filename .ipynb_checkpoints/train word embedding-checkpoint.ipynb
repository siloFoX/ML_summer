# Word embeddings 과 여러기법들

시각화 툴 : [Embedding Projector](http://projector.tensorflow.org)

word embedding 이란 단어(word)를 R차원의 vector로 매칭시켜주는 작업임.

##### 물론 단어가 아닌 다른 어떤 것을 vectorize할 수도 있겠죠?

## 1. one-hot encoding

어휘 하나를 index 한개에 배치하는 법

밑에 그림대로.
<img src="https://github.com/tensorflow/docs/blob/master/site/en/r2/tutorials/text/images/one-hot.png?raw=1" alt="Diagram of one-hot encodings" width="400" />
비효율적인 점 : 단어의 개수가 많아질 수록 1인 데이터가 희소해진다.(sparse vector) 그 말은 데이터의 대부분이 0으로 되어있다는 뜻

## 2. Encode each word with a number

고유한 숫자를 한 단어랑 짝짓는 것 예를 들어 "cat"에 1을, "dog"에 2를 할당하는 것이다.
그런 식으로 [3, 1, 2, 4, 1] 이런 고밀도의 vector에 인코딩할 수 있다.

우선 효율적이다 sparse vector에 비해서. 

단점
- 정수 인코딩은 서로 임의적이다.(단어 사이의 관계가 안 담겨있다)
- model이 해석하기 어려워할 수 있다. 원 핫을 왜 썼는지 생각해보면 알 수 있다. 학습에 그닥 효율이 없다는 뜻이다.

## 3. word embeddings

단어들 마다 유사한 포멧의 인코딩을 써서 나타낸다.(일정한 크기의 벡터 안에 있는 값을 조정해서 고유한 단어를 나타낸다.)

장점은 우리가 직접 인코딩을 지정할 필요가 없다는 점이다. floating point 값으로 나타낸다. embedding하는 것 자체가 모델이므로(값을 부여해주는 모델을 만들어야 한다는 뜻) weight를 training 시켜야 한다.

주로 8차원에서 1024차원의 임베딩을 쓰고 차원이 늘어날 수록 당연히 데이터가 많이 필요하다.

<img src="https://github.com/tensorflow/docs/blob/master/site/en/r2/tutorials/text/images/embedding2.png?raw=1" alt="Diagram of an embedding" width="400"/>

위 그림은 4차원의 floating point 벡터로 표현되었다.

# Embedding layer

````python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
````

embedding_layer = layers.Embedding(input_dim = 1000, output_dim = 32)

1000개의 단어를 32차원의 벡터로 표현하는 층임

````python
type(embedding_layer)
````

1000개의 단어를 영화리뷰글에서 추출해냈다.(케라스 예제)

단어에 따른 감정을 분류해보자

````python
vocab_size = 10000

imdb = keras.datasets.imdb

(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words = vocab_size)
````

각 정수는 dictionary에서 특정 단어의 인덱스임(2번으로 인코딩이 되어있음)

````python
train_data[0]
````

````python
word_index = imdb.get_word_index()
````

그래서 단어랑 매칭시켜주려고 dictionary를 불러왔다

````python
word_index
````

value를 3씩 더해줬다 단어 말고 다른 처리 값이 필요하기 때문에(데이터가 원래 그렇게 쓰도록 권장되어 있다;;;) 

0부터 3까지의 value는 비워둔 것이다

````python
word_index = {k:(v+3) for k,v in word_index.items()}
````

````python
word_index
````

````python
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2  # unknown
word_index["<UNUSED>"] = 3
````

index로 search할 것 이므로 key와 value 값의 위치를 switch 한다

````python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
````

확인한다

````python
reverse_word_index
````

decoding 하는 function을 만든다

````python
def decode(text) :
    return ' '.join([reverse_word_index.get(i, '<ERR>') for i in text])
````

````python
decode(train_data[101])
````

최대길이를 500자로 정하고 넘으면 자르고 적으면 padding(zero-post-padding)한다(pre-padding도 있다)

````python
maxlen = 500

train_data = keras.preprocessing.sequence.pad_sequences(train_data,
                                                        value = word_index["<PAD>"],
                                                        padding = 'post',
                                                        maxlen = maxlen)

test_data = keras.preprocessing.sequence.pad_sequences(test_data,
                                                       value = word_index["<PAD>"],
                                                       padding = 'post',
                                                       maxlen = maxlen)
````

````python
train_data[0].shape
````

````python
train_data[0].shape
````

model에 관한 간단한 설명을 해주겠다

오프라인에서만 해줄거지롱

````python
embedding_dim = 16

model = keras.Sequential([
  layers.Embedding(vocab_size, embedding_dim, input_length = maxlen),
  layers.GlobalAveragePooling1D(),
  layers.Dense(16, activation = 'relu'),
  layers.Dense(1, activation = 'sigmoid')
])

model.summary()
````

````python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

history = model.fit(train_data, train_labels, epochs = 30, batch_size = 512, 
                    validation_data = (test_data, test_labels))
````

embedding이 잘 되게 하려면 전체적인 모델의 성능이 좋아야 한다

````python
import matplotlib.pyplot as plt

history_dict = history.history

acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)

plt.figure(figsize=(12,9))
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.figure(figsize=(12,9))
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim((0.5,1))
plt.show()
````

embedding layer만 따로 뽑아낸다

````python
embedding_layer = model.layers[0]
embedding_weights = embedding_layer.get_weights()[0]
print(embedding_weights.shape)
````

````python
import io

out_v = io.open('vecs.tsv', 'w', encoding='utf-8')
out_m = io.open('meta.tsv', 'w', encoding='utf-8')

for word_num in range(vocab_size):
    word = reverse_word_index[word_num]
    embeddings = embedding_weights[word_num]
    out_m.write(word + "\n")
    out_v.write('\t'.join([str(x) for x in embeddings]) + "\n")
    
out_v.close()
out_m.close()
````

[Embedding Projector](http://projector.tensorflow.org/) 왼쪽을 눌러라.

* "Load data" 를 눌러라.

* ```vecs.tsv``` 와 ```meta.tsv```를 업로드 해라. 

````python
RNN_model = tf.keras.Sequential([
    layers.Embedding(vocab_size, 64),
    layers.Bidirectional(tf.keras.layers.LSTM(64)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
````

````python
RNN_model.summary()
````

````python
RNN_model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
````

````python
history = RNN_model.fit(train_data, train_labels, epochs = 10,
                    validation_data = (test_data, test_labels))
````

````python
history_dict = history.history

acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)

plt.figure(figsize=(12,9))
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.figure(figsize=(12,9))
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim((0.5,1))
plt.show()
````
